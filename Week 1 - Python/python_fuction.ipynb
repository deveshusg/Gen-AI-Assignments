```python
# Write a python fuction which takes "Text, chunk size, overlap size" as argument to divide the given text into chunks of given size with given overlap size


text=input("Enter the text: ")
chunk=int(input("Enter the chunk size: "))
overlap=int(input("Enter the overlap size: "))
def text_chunk_overlap(text,chunk,overlap):
    step=chunk-overlap
    for i in range(0,len(text),step):
        print(text[i:i+chunk])

text_chunk_overlap(text,chunk,overlap)
```
